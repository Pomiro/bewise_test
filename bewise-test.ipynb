```python
# Импорт библиотек
import pandas as pd
import re

# Настройки для лучшего отображение данных в таблице
pd.set_option('display.max_columns', None)
pd.set_option('display.expand_frame_repr', False)
pd.set_option('max_colwidth', -1)

# Загружаем данные
data = pd.read_csv('данные')
# Имена сотрудников
names = ['анастасия', 'ангелина', 'максим']

# Создание функции. data - таблица с данными; names - имена сотрудников                   
def dialogue(data, names):
    # Создание таблицы для вывода результа скрипта
    dialogue_df = pd.DataFrame(columns=['dlg_id','Приветствие', 'Представление', 'Имя', 'Компания', 'Завершение', 'Требование']).set_index('dlg_id')  
    # Цикл по всем диалогам в данных
    for f in range(data.dlg_id.nunique()):
        # Берем из данных только реплики сотрудника и один из диалогов
        dlg = data[data.role == 'manager'][data.dlg_id == f].reset_index(drop=True) 
        # Вводим переменные
        hi = ''
        bb = ''
        dating = ''
        name = ''
        company = ''
        # Находим приветствие сотрудника
        for i in range(len(dlg)):
            if hi == '' and re.findall(r'добрый|здравствуйте', dlg.text[i]):
                hi = dlg.text[i]
        # Находим прощение сотрудника
        for i in range(len(dlg)):
            if re.findall(r'свидания|всего', dlg.text[i]):
                bb = dlg.text[i]
        # Находим представление сотрудника и его имя
        for i in range(len(dlg)):
            for k in names:
                if name == '' and re.findall(fr'{k}', dlg.text[i]) != []:
                    dating = dlg.text[i]
                    name = k
        # Находим название компании
        for i in range(len(dlg)):
            if company == '' and re.findall(r'компания', dlg.text[i]):
                company = re.findall(r'компания \w+', dlg.text[i])[0].split(" ")[1]
        # Проверка требование на приветствие и прощание
        if hi == '' or bb == '':
            civility = 'Нет'
        else: 
            civility = 'Да'
        # Формируем полученные данные и формируем таблицу
        new_row = pd.Series(data={'Приветствие':hi, 'Представление':dating, 'Имя':name, 'Компания':company, 'Завершение':bb, 'Требование':civility}, name=f)
        dialogue_df = dialogue_df.append(new_row, ignore_index=False)
    return(dialogue_df)
dialogue(data, names)
```
